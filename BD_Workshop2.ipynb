```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df = df.drop_duplicates()
```

```python
df.hist(figsize=(12, 14))
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.countplot(x='gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['study_hours_per_day'], bins=10)
plt.title("Distribution of Study Hours")
plt.xlabel("Study Hours")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x=df['phone_usage_hours'])
plt.title("Boxplot of Screen Time")
plt.show()
```

```python
sns.scatterplot(x='study_hours_per_day', y='productivity_score', data=df)
plt.title("Scatterplot of Study Hours vs Productivity Score")
plt.show()
```

```python
sns.scatterplot(x='phone_usage_hours', y='productivity_score', data=df)
plt.show()
```

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(20,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(40, 12))
sns.boxplot(data=df)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.histplot(df['productivity_score'], kde=True)
plt.title("Distribution of Productivity Score")
plt.show()
```
